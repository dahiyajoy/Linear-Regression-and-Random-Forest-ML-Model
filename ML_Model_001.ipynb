# **Linear Regression and Random Forest ML Model**

# **Loading Dataset**

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

# **Data Preparation**

## Data Separation as x and y

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

## Data Splitting


```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=100)
```

```python
y_train
```

# **Model Building**

## **Linear Regression**


### **Training the Model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

### **Applying the Model to make Prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
 y_lr_train_pred
```

### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression',lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Model','Train MSE','Train R2','Test MSE', 'Test R2']
lr_results
```

## **Random Forest**

### **Training the Model**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

### **Applying the Model to make Prediction**

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest',rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Model','Train MSE','Train R2','Test MSE', 'Test R2']
rf_results
```

## **Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models.reset_index(drop=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a bias term with ones
X_b = np.c_[np.ones((100, 1)), X]

# Calculate the optimal weights using the Normal Equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print(f"Optimal weights (theta): {theta_best}")

# Make predictions
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)

# Plot the results
plt.plot(X_new, y_predict, "r-", label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
